```python
import pandas as pd
```

```python
import pyarrow.parquet as pq
```

```python
from sqlalchemy import create_engine
```

```python
from tqdm.auto import tqdm
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
# Read parquet directly from URL
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet')

# Create table
df.head(0).to_sql(
    name="green_taxi_data",
    con=engine,
    if_exists="replace"
)
print("Table created")

# Insert data
df.to_sql(
    name="green_taxi_data",
    con=engine,
    if_exists="append"
)
print("Inserted rows:", len(df))
```

```python
df = pd.read_csv('https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv')

# Create table
df.head(0).to_sql(
    name="taxi_zone_data",
    con=engine,
    if_exists="replace"
)
print("Table created")

# Insert data
df.to_sql(
    name="taxi_zone_data",
    con=engine,
    if_exists="append"
)
print("Inserted rows:", len(df))
```
